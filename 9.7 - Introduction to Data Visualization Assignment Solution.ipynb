#### **Problem 1**

#### Plot the total profit across the months

* X label name = Month Number
* Y label name = Total profit



```
# Your code here

company_profit = company_sales['total_profit']
months = company_sales['month_number']
```

```
plt.figure(figsize = (10,7))
plt.plot(months,company_profit, 'bo-')
plt.title('Profit Per Month Of A Company ', {'fontsize' : 15, 'fontweight' : 30})
plt.xlabel('Month Number', {'fontsize' : 12, 'fontweight' : 20})
plt.ylabel('Total profit', {'fontsize' : 12, 'fontweight' : 20})
plt.show()
```

#### <b>Problem 2 </b>

#### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
number_of_units_sold = company_sales.groupby(['month_number'])[['facewash','facecream']].sum()
ax = number_of_units_sold.plot.bar(width=0.8, linewidth=1, edgecolor='black', figsize = (15,8))

ax.set_title("Face Cream and Facewash Product Sales Month Wise", fontsize = 20)
ax.set_xlabel("Month Number", fontsize = 15)
ax.set_ylabel("Product Sales", fontsize = 15)
plt.show()
```

#### <b>Problem 3</b>

#### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here

units_sold = company_sales.groupby(['month_number'])['facecream'	,'facewash'	,'toothpaste'	,'bathingsoap',	'shampoo'	,'moisturizer'].sum()

plt.figure(figsize = (15,8))
plt.plot(units_sold)
plt.title('Sum Of All The Units SOld Per Month', {'fontsize' : 15, 'fontweight' : 30})
plt.legend(['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.xlabel('Month Number', {'fontsize' : 12, 'fontweight' : 20})
plt.ylabel('Sum Of Units Per Month', {'fontsize' : 12, 'fontweight' : 20})
plt.show()

# WE CANNOT SEE FACEWASH IN THE PLOT IS BECAUSE FACEWASH AND MOISTURISER HAS THE SAME VALUES
```
